```python
import pandas as pd
import numpy as np
import ast
```

```python
movies=pd.read_csv("movies_metadata.csv")
credit=pd.read_csv("credits.csv")
```

```python
movies.head()
movies["id"]
```

```python
credit.head(50)
```

```python
movies_df=pd.concat([movies,credit],axis=1)
movies_df.head()
```

```python
movies_df=movies_df[["id","title","overview","genres","cast","crew"]]
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df.isnull().sum()
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df.isnull().sum()
```

```python
movies_df.duplicated().sum()
```

```python
movies_df.drop_duplicates()
```

```python
movies_df.duplicated().sum()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies_df["genres"]=movies_df["genres"].apply(convert)
movies_df.head()
```

```python
def convert_cast(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<3:
            L.append(i["name"])
            count+=1
        else:
            break
    return L    
```

```python
movies_df["cast"]=movies_df["cast"].apply(convert_cast)
```

```python
movies_df.head()
```

```python
movies_df["crew"].values[0]
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
    return L        
```

```python
movies_df["crew"]=movies_df["crew"].apply(convert_crew)
```

```python
movies_df["crew"].values[0]
```

```python
movies_df["overview"]=movies_df["overview"].apply(lambda x:x.split())
```

```python
movies_df.head()
```

```python
movies_df["overview"]=movies_df["overview"].apply(lambda x:[i.replace(" ","")for i in x])
movies_df["genres"]=movies_df["genres"].apply(lambda x:[i.replace(" ","")for i in x])
#movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","")for i in x])
movies_df["cast"]=movies_df["cast"].apply(lambda x:[i.replace(" ","")for i in x])
movies_df["crew"]=movies_df["crew"].apply(lambda x:[i.replace(" ","")for i in x])
movies_df.head()
```

```python
movies_df["tags"]=movies_df["overview"]+movies_df["genres"]+movies_df["cast"]+movies_df["crew"]
movies_df.head()
```

```python
new_df=movies_df[["id","title","tags"]]

new_df.head()
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"]=new_df["tags"].str.lower()
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
        
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"].values[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
vectors
```

```python
vectors.shape
```

```python
features=cv.get_feature_names_out()
features
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie_name:str):
    movie_index=new_df[new_df["title"]==movie_name].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend("Spectre")
```
